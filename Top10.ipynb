```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("311_Service_Requests_2020.csv") #read dataset
df.head()
```

```python
df_10027 = df[df['Incident Zip']==10027] #get info for 10027 zip code
df_10027.head()
```

```python
top10 = df_10027.groupby('Complaint Type').size().sort_values(ascending=False).head(10)
#calculate incidents number in each complaint type, sort it descendingly and display top 10
```
